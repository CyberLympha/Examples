# **Кластеризация методом KNN**

# Описание метода

kNN (расшифровывается как k Nearest Neighbor или k Ближайших Соседей) — это один из самых простых алгоритмов классификации, также иногда используемый в задачах регрессии.

В алгоритме вычисляется расстояние от новой точки данных до всех других обучающих точек данных. Расстояние может быть любого типа, например евклидово или манхэттенское. Затем выбираетсят K-ближайшие точки данных, где K может быть любым целым числом. Наконец, точка данных присваивается классу, к которому принадлежит большинство из K точек данных.

**Алгоритм**

Для классификации каждого из объектов тестовой выборки необходимо последовательно выполнить следующие операции:
  - Вычислить расстояние до каждого из объектов обучающей выборки
  - Отобрать k объектов обучающей выборки, расстояние до которых минимально
  - Класс классифицируемого объекта — это класс, наиболее часто встречающийся среди k ближайших соседей

**Гиперпараметры**

KNN включает два гиперпараметр: значение K и функцию расстояния.

Значение K определяет количество соседей, участвующих в алгоритме kNN.  k должен быть настроен на основе ошибки проверки.
Чаще всего используется евклидово расстояние как функция расстояний, но есть и другие альтернативы: манхэттенское расстояние, расстояние Хемминга, расстояние Минковского. В контексте данных микроматрицы экспрессии генов, например, k -NN использовался с коэффициентами корреляции, такими как Пирсон и Спирмен, в качестве показателя. Часто точность классификации k -NN может быть значительно улучшена, если метрика расстояния изучена с помощью специализированных алгоритмов, таких как анализ компонентов ближайшего соседа с большим запасом или соседства.

Точность алгоритма k -NN может быть сильно снижена присутствием зашумленных или нерелевантных функций или несоответствием масштабов функций и их важности. Особенно популярный для улудшения классификации подходом является использованием эволюционных алгоритмов для оптимизации особенности масштабирования. Другой популярный подход - масштабирование функций за счет взаимной информации обучающих данных с обучающими классами. То есть примеры более частого класса имеют тенденцию доминировать в предсказании нового примера, потому что они имеют тенденцию быть общими среди k ближайших соседей из-за их большого количества. Одним из способов решения этой проблемы является взвешивание классификации с учетом расстояния от контрольной точки до каждого из ее k ближайших соседей. Класс (или значение в задачах регрессии) каждой из k ближайших точек умножается на вес, пропорциональный обратной величине расстояния от этой точки до контрольной точки. Другой способ преодоления перекоса - абстракция в представлении данных. 

Наилучший выбор k зависит от данных; как правило, более высокие значения k уменьшают влияние шума на классификацию, но делают границы между классами менее четкими. Хороший k можно выбрать с помощью различных эвристических методов (см. Оптимизация гиперпараметров ). Частный случай, когда прогнозируется, что класс является классом ближайшей обучающей выборки (т.е. когда k = 1), называется алгоритмом ближайшего соседа. В задачах двоичной (двухклассовой) классификации полезно выбрать k как нечетное число, так как это позволяет избежать равных голосов. Одним из популярных способов выбора эмпирически оптимального k в этой настройке является метод начальной загрузки.

**Свойства**

k -NN - это частный случай "баллонной" оценки плотности ядра с переменной полосой пропускания и однородным ядром.

Наивную версию алгоритма легко реализовать, вычислив расстояния от тестового примера до всех сохраненных примеров, но это требует больших вычислительных ресурсов для больших данных. Использование приближенного алгоритма поиска ближайшего соседа делает k- NN вычислительно управляемым даже для больших наборов данных. Было предложено много алгоритмов поиска ближайшего соседа; они обычно стремятся сократить количество фактически выполняемых оценок расстояния.

k- NN имеет некоторые сильные результаты согласованности . Поскольку количество данных приближается к бесконечности, двухклассовый алгоритм k- NN гарантированно дает частоту ошибок не хуже, чем удвоенную частоту ошибок Байеса (минимально достижимую частоту ошибок с учетом распределения данных). Различные улучшения скорости k -NN возможны с помощью графов близости.

**Абстрактный алгоритм работы метода kNN**

Алгоритм K-ближайших соседей (KNN) использует «сходство признаков» для прогнозирования значений новых точек данных, что также означает, что новой точке данных будет присвоено значение на основе того, насколько близко он соответствует точкам в обучающем наборе. Мы можем понять его работу с помощью следующих шагов —

*Шаг 1*. Загрузить данные для обучения

*Шаг 2*. Ввести значение гиперпараметра k
 
*Шаг 3* . Для каждой точки в тестовых данных сделать следующее:

  - 3.1 — Рассчитайть расстояние между данными испытаний и каждой строкой данных тренировки с помощью любого из методов: Евклидово, Манхэттенское расстояние или расстояние Хэмминга. 

  - 3.2 — Отсортировать расстояния в порядке возрастания.

  - 3.3 — Выбрать верхние k строк из отсортированного массива.

  - 3.4 — Назначить класс точке на основе наиболее часто встречающегося класса этих строк.


**Сложность алгоритма**

Сложность обучения - $O(1)$. Технически правильным ответом также является $O(N)$, так как требуется запоминать обучающую выборку.

Сложность прогнозирования - $O(N)$ для каждого объекта. Если по фиксированной обучающей выборке требуется независимо сделать прогноз для $k$ объектов, сложность $O(kN)$.

**Преимущества и недостатки данного метода кластеризации**

Преимущества: 

  - Это очень простой алгоритм для понимания и интерпретации.

  - Это очень полезно для нелинейных данных, потому что в этом алгоритме нет предположения о данных.

  - Это универсальный алгоритм, поскольку мы можем использовать его как для классификации, так и для регрессии.

  - Он имеет относительно высокую точность, но есть гораздо лучшие контролируемые модели обучения, чем KNN.

Недостатки:

  - Это вычислительно немного дорогой алгоритм, потому что он хранит все данные обучения.

  - Требуется большой объем памяти по сравнению с другими контролируемыми алгоритмами обучения.

  - Прогнозирование медленное в случае большого N.

  - Он очень чувствителен к масштабу данных, а также к несущественным функциям.

**Сравнение с другими алгоритмами**

Общая разница между KNN и другими моделями заключается в больших вычислениях в реальном времени, необходимых для KNN по сравнению с другими.

КНН против наивного Байеса:

  - Наивный байес намного быстрее чем KNN из-за выполнения KNN в реальном времени.
  - Наивный байесовский параметр является параметрическим, тогда как KNN является непараметрическим.

КНН против линейной регрессии:

  - KNN лучше, чем линейная регрессия, когда данные имеют высокий SNR.

КНН против СВМ:

  - SVM заботится о выбросах лучше, чем KNN.
  - Если данные обучения намного больше, чем нет. из особенностей (m >> n), KNN лучше, чем SVM. SVM выигрывает у KNN, когда есть большие возможности и меньше обучающих данных.

КНН против нейронных сетей:

  - Нейронные сети нуждаются в больших данных обучения по сравнению с KNN для достижения достаточной точности.
  - NN нуждается в большом количестве настроек гиперпараметров по сравнению с KNN.



# Примеры применения

Алгоритм KNN может применяться практически во всех задачах классификации, особенно в тех случаях, когда оценить параметры вероятностного распределения данных сложно или невозможно. 

Данный алгоритм является одним из простейших алгоритмов классификации, так что в реальных задачах его применимость не так уж и велика, как могло бы показаться, однако он отлично подходит для использования в учебных целях на искусственных данных.

**Применение данного алгоритма:** 

Классификация клиентов (например, по уровню лояльности);

*Медицина* — классификация пациентов по медицинским показателям;

*Маркетинг* — классификация товаров по уровню популярности;

*Банковская система* - KNN может использоваться, чтобы предсказать, подходит ли человек для утверждения кредита, есть ли у этого человека характеристики, аналогичные неплательщикам;

*Политика* - с помощью алгоритмов KNN мы можем классифицировать потенциального избирателя по разным классам, таким как «Буду голосовать», «Не буду голосовать», «Буду голосовать за партию« Конгресс», « Буду голосовать за партию «BJP».

Другими областями, в которых может использоваться алгоритм KNN, являются распознавание речи, обнаружение почерка, распознавание изображений и распознавание видео.

**Построение модели**

Нам нужно определить переменную «X» и переменную «Y», чтобы построить нашу модель KNN. Учитывая наш набор данных, нам нужно классифицировать переменную «species», поэтому ее можно использовать как переменную «Y» или как зависимую переменную. Все остальные переменные в нашем наборе данных можно рассматривать как независимые переменные или переменные «X».

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
y = df['species'].values
```

Нормализуем значения переменной «X», которые могут быть полезны при обучении нашей модели KNN (нормализация - это процесс структурирования и удаления аномалий для заданного набора значений). Это также снижает количество данных и улучшает целостность данных. Для нормализации используем функцию $StandardScaler$ в $scikit-learn$.

```python
X = StandardScaler().fit(X).transform(X.astype(float))
```

Разделим данные на обучающую часть и тестовую, где обучающий набор имеет наибольшее количество точек данных. В обучающую часть определеим 75% данных. Чтобы разделить наши данные, используем функцию $train_test_split$, предоставляемую $scikit-learn$ в $python$.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

Переходим к реализации KNN. Обучим модель на тестовой части c помощью метожа $.fit$. Видим, что по умолчанию использовано расстояние Минковского, но существуют и другие описанные ранее варианты, которые следует определять вручную. 

```python
k = 3
KNN = KNeighborsClassifier(n_neighbors = k)
KNN.fit(X_train, y_train)
```

Завершив обучение модели, пробуем спрогнозировать ответ на тестовых данных с помощью метода $.predict$.

```python
Y = KNN.predict(X_test)
```

Проверим, насколько точным оказался прогноз модели. 

```python
print(round(accuracy_score(y_test, Y)*100, 2))
```

Точность предсказания модели превысила 90%, что является отличным результатом.

Выбранное случайно число 3 в качестве значения $k$ привело к высокой точности.

Один из способов помочь вам найти наилучшее значение $k$ - построить график значения K и соответствующей частоты ошибок для набора данных.

Построим среднюю ошибку для прогнозируемых значений тестового набора для всех значений $k$ в диапазоне от 1 до 100. Для этого сначала вычислим среднее значение ошибки для всех прогнозируемых значений, где $k$ колеблется от 1 до 100. Затем остроим график значений error к значениям $k$

```python
# строим среднюю ошибку
error = []
for i in range(1, 100):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))

# строим график   
plt.figure(figsize=(20, 10))
plt.plot(range(1, 100), error, color='blue', linestyle='dashed', marker='o',
         markerfacecolor='green', markersize=8)
plt.title('Частота ошибок при значении k')
plt.xlabel('Значение k')
plt.ylabel('Средняя ошибка')
```

По графику видно, что выбранное изначально значение $k$ является оптимальным.
