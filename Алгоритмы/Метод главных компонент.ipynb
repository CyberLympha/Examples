<h2><strong>Метод главных координат (PCA - Principal component analysis)</strong></h2>

---
Метод главных компонент (PCA) — один из основных способов уменьшить размерность данных, потеряв наименьшее количество информации.
Изобретён Карлом Пирсоном в 1901 году.


---
<h3><strong>Описание алгоритма</strong></h3>


---
Мы берём матрицу данных X, имеющую размерность NxD, и умножаем её на матрицу преобразования Q, имеющую размерность DxD, в результате чего получаем матрицу преобразованных данных Z, также имеющую размерность NxD:

Z=XQ. 

Если же мы захотим преобразовать отдельный вектор x в соответствующий отдельный вектор z, то умножение приобретает вид

Z=Qx. 

Изменение порядка умножения связано с тем, что матрица данных X в данном случае представляет собой вектор-строку размерностью 1xD, тогда как мы, говоря об отдельных векторах, имеем в виду вектор-столбец размерностью Dx1.

---
<h3><strong>Применение метода</strong></h3>


---
Применяется для визуализации данных (является одним из самых широко используемых методов), компрессии изображения и видео (в данной области метод используется не так активно из-за большой вычислительной сложности), подавления шума на изображениях, индексации видео, биоинформатики, хемометрики, психодиагностики, эконометрики и сокращения размерности динамических моделей.

---
<h3><strong>Реализация метода на Python</strong></h3>





```python
#Импорт библиотек
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Создание набора данных
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

```python
#Импортируем метод и используем его для уменьшения размерности нашего датасета
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
pca.fit(X)
X_new = pca.transform(X)
print("Изначальная размерность:   ", X.shape)
print("Измененная размерность после применения метода:", X_new.shape)
```

```python
#Демонстрация на графике изменения размерности
#Для этого мы преобразуем трансформированные данные обратно к размерности 2
X_inverse = pca.inverse_transform(X_new)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_inverse[:, 0], X_inverse[:, 1], alpha=0.8)
plt.axis('equal');
#Оранжевые точки - преобразованные данные методом главных компонент
#Прозрачные голубые - изначальные данные
```
