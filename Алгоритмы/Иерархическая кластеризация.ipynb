## Понятие кластеризации

Кластеризация (или кластерный анализ) — это задача разбиения множества объектов на группы, называемые кластерами. Внутри каждой группы должны оказаться «похожие» объекты, а объекты разных группы должны быть как можно более отличны. Главное отличие кластеризации от классификации состоит в том, что перечень групп четко не задан и определяется в процессе работы алгоритма.

Применение кластерного анализа в общем виде сводится к следующим этапам:
1. Отбор выборки объектов для кластеризации.
2. Определение множества переменных, по которым будут оцениваться объекты в выборке. При необходимости – нормализация значений переменных.
3. Вычисление значений меры сходства между объектами.
4. Применение метода кластерного анализа для создания групп сходных объектов (кластеров).
5. Представление результатов анализа.

После получения и анализа результатов возможна корректировка выбранной метрики и метода кластеризации до получения оптимального результата.


## Иерархическая кластеризация

Иерархическая кластеризация — множество алгоритмов кластеризации, направленных на создание иерархии вложенных разбиений исходного множества объектов. Методы иерархической кластеризации можно условно поделить на две большие категории: агломеративные и дивизимные (разделительные). Агломеративные объединяют мелкие кластеры в более крупные и отображают историю этих объединений в виде специального графического представления, называемого дендрограммой (деревоподобное изображение иерархической кластерной структуры, подробнее о них ниже). Дивизимные методы, наоборот, основаны на том, что в конечную выборку объектов мы по очереди дробим на всё более и более мелкие кластеры.

## Метод агломеративной иерархической кластеризации
Алгоритм метода следующий:
1. Принять каждую точку данных за отдельный кластер, тем самым сформировать N кластеров.
2. Взять две ближайшие точки данных и объединить их в один кластер, то есть сформировать N-1 кластер
3. Взять два ближайших кластера и объединить их в один, сформировать N-2 кластера
4. Повторять шаг 3, пока не останется только один кластер

Чтобы лучше понять, давайте рассмотрим наглядное представление метода агломеративной иерархической кластеризации. Допустим,у нас есть шесть точек данных {A, B, C, D, E, F}.


Шаг 1: На начальном этапе мы вычисляем близость отдельных точек и рассматриваем все шесть точек данных как отдельные кластеры, как показано на рисунке ниже.


image.png

Шаг 2: На втором этапе подобные кластеры объединяются вместе и формируются как один кластер. Рассмотрим B,C и D,E-подобные кластеры, которые объединяются на втором шаге. Теперь у нас осталось четыре кластера, которыми являются A, BC, DE, F.

Шаг 3: Мы снова вычисляем близость новых кластеров и объединяем аналогичные кластеры, чтобы сформировать новые кластеры A, BC, DEF.

Шаг 4: Вычислите близость новых кластеров. Кластеры DEF и BC похожи и объединены вместе, чтобы сформировать новый кластер. Теперь у нас осталось два кластера A, BCDEF.

Шаг 5: Наконец, все кластеры объединяются вместе и образуют единый кластер.

## Метод дивизимной иерархической кластеризации

Разделительная иерархическая кластеризация является полной противоположностью агломеративной иерархической кластеризации. При разделительной иерархической кластеризации мы рассматриваем все точки данных как один кластер и на каждой итерации отделяем точки данных от кластера, которые не похожи. Каждая выделенная точка данных рассматривается как отдельный кластер. В конце концов, мы останемся с n кластерами.
Поскольку мы делим отдельные кластеры на n кластеров, это называется разделительной иерархической кластеризацией.

## Вычисление расстояния между кластерами

Вычисление расстояния между двумя кластерами важно для объединения или разделения кластеров. Существуют определенные подходы, которые используются для расчета расстояния между двумя кластерами:
1. MIN (single-linkage algorithm)
2. MAX (complete linkage algorithm)
3. Group average
4. Distance between centroids
5. Ward's Method




### MIN (single-linkage algorithm)

Согласно этому подходу расстояние между двумя кластерами C1 и С2 - это минимальное расстояние между точками Pi и Pj такими, что Pi принадлежит С1, а Pj принадлежит C2.

Проще говоря, выберите две ближайшие точки таким образом, чтобы одна точка находилась в первом кластере, а другая во втором кластере, возьмите расстояние между ними и объявите его как расстояние между двумя кластерами.

С помощью такого подхода можно разделять неэллиптические формы до тех пор, пока расстояние между ними не будет слишком маленьким, однако такой подход не работает при наличии шума.

### MAX (complete linkage algorithm)

Полная противоположность подходу MIN. Согласно этому подходу расстояние между двумя кластерами C1 и С2 - это максимальное расстояние между точками Pi и Pj такими, что Pi принадлежит С1, а Pj принадлежит C2.

Проще говоря, выберите две наиболее дальние друг от друга точки таким образом, чтобы одна точка находилась в первом кластере, а другая во втором кластере, возьмите расстояние между ними и объявите его как расстояние между двумя кластерами.

### Group average

Принцип средней связи заключается в следующем: нужно взять все пары точек, где одна из них принадлежит кластеру С1, а другая - кластеру С2, посчитать расстояния между ними и взять среднее этих расстояний в качестве расстояния между кластерами.

### Distance between centroids (расстояние между центроидами)
Центроид - это точка, полученная усреднением положений всех точек в кластере.

Суть такого подхода заключается в поиске расстояний между цетроидами кластеров и принятием этих расстояний за расстояния между кластерами.


image.png

### Ward’s Method (метод Уорда)

Данный подход очень похож на принцип средней связи, однако здесь берется квадрат расстояния между точками кластеров, а уже потом высчитывается среднее.

## Емкостная и временная сложность алгоритмов иерархической кластеризации

### Емкостная сложность

Объем памяти, необходимый для реализации метода иерархической кластеризации напрямую зависит от количества точек данных, поскольку нужно хранить в памяти квадратную матрицу сходства. Значит, емкостная сложность $O(n^2)$, где n - количество точек данных. 

### Временная сложность

Поскольку мы должны выполнить n итераций и на каждой итерации нам нужно обновлять матрицу сходства и восстанавливать матрицу, временная сложность также очень высока. Временная сложность равна $O(n^3)$, где n - количество точек данных.

#### Элементы дендрограммы

image.png

Дендрограммы могут быть разной формы, но как правило, они являются столбчатым графом (как на рисунках выше) или строковым графом. Независимо от формы, базовый график состоит из одних и тех же частей:
1. The Clades (клайды) - ветви, они расположены в соответствии с тем, насколько похожи (или непохожи). Клайды, близкие к одной и той же высоте, похожи друг на друга; клайды с разной высотой отличаются друг от друга — чем больше разница в высоте, тем больше различия.
2. У каждого клайда один или более **листьев**. Теоретически у клайда может быть бесконечное количество листьев. Однако чем больше у вас листьев, тем сложнее будет прочитать график невооруженным глазом.
3. Листья A, B и C больше похожи друг на друга, чем на листья D, E или F.
4. Листья D и E больше похожи друг на друга, чем на листья A, B, C или F.
5. Лист F существенно отличается от всех остальных листьев.

## Примеры на python

### Пример 1

Давайте реализуем один сценарий использования с использованием алгоритма агломеративной иерархической кластеризации. Набор данных состоит из сведений о клиентах одного конкретного торгового центра вместе с их счетом расходов.

Вы можете скачать набор данных с сайта здесь: https://gist.github.com/ryanorsinger/cb1222e506c1266b9cc808143ddbab82#file-mall_customers-csv.

Давайте начнем с импорта 3 основных библиотек:

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

Загружаем набор данных:

```python
dataset = pd.read_csv('./mall_customers.csv')
dataset.head()
```

Таким образом, наша цель - кластеризация клиентов на основе их расходов. Поля customer_id и gender не являются значимыми и могут быть опущены. Чтобы создать матрицу независимых переменных возьмем поля age и annual_income. 

```python
X = dataset.iloc[:, [3, 4]].values
```

Далее нужно выбрать число кластеров. Для этого воспользуемся дендрограммой.

```python
import scipy.cluster.hierarchy as sch
dendrogrm = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
plt.show()
```

Чтобы выбрать количество кластеров, мы проводим горизонтальную линию через самую длинную вертикальную линию, которая проходит максимальное расстояние вверх и вниз, не пересекая точки слияния. Количество линий вертикали, которые пересекает нарисованная нами горизнтальная линия, является оптимальным количеством кластеров.

В данном случае это 5. Итак, давайте сопоставим нашу агломеративную модель с 5 кластерами.

```python
from sklearn.cluster import AgglomerativeClustering 
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward') 
y_hc = hc.fit_predict(X)

# Visualising the clusters 
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 50, c = 'red', label = 'Careful') 
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 50, c = 'blue', label = 'Standard') 
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 50, c = 'green', label = 'Target') 
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 50, c = 'cyan', label = 'Careless') 
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 50, c = 'magenta', label = 'Sensible') 
plt.title('Clusters of customers') 
plt.xlabel('Annual Income (k$)') 
plt.ylabel('Spending Score (1-100)') 
plt.legend() 
plt.show()
```

### Пример 2

```python
print(__doc__)
from time import time

import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt

from sklearn import manifold, datasets

X, y = datasets.load_digits(return_X_y=True)
n_samples, n_features = X.shape

np.random.seed(0)

def nudge_images(X, y):
    # Having a larger dataset shows more clearly the behavior of the
    # methods, but we multiply the size of the dataset only by 2, as the
    # cost of the hierarchical clustering methods are strongly
    # super-linear in n_samples
    shift = lambda x: ndimage.shift(x.reshape((8, 8)),
                                  .3 * np.random.normal(size=2),
                                  mode='constant',
                                  ).ravel()
    X = np.concatenate([X, np.apply_along_axis(shift, 1, X)])
    Y = np.concatenate([y, y], axis=0)
    return X, Y


X, y = nudge_images(X, y)


#----------------------------------------------------------------------
# Visualize the clustering
def plot_clustering(X_red, labels, title=None):
    x_min, x_max = np.min(X_red, axis=0), np.max(X_red, axis=0)
    X_red = (X_red - x_min) / (x_max - x_min)

    plt.figure(figsize=(6, 4))
    for i in range(X_red.shape[0]):
        plt.text(X_red[i, 0], X_red[i, 1], str(y[i]),
                 color=plt.cm.nipy_spectral(labels[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    plt.xticks([])
    plt.yticks([])
    if title is not None:
        plt.title(title, size=17)
    plt.axis('off')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

#----------------------------------------------------------------------
# 2D embedding of the digits dataset
print("Computing embedding")
X_red = manifold.SpectralEmbedding(n_components=2).fit_transform(X)
print("Done.")

from sklearn.cluster import AgglomerativeClustering

for linkage in ('ward', 'average', 'complete', 'single'):
    clustering = AgglomerativeClustering(linkage=linkage, n_clusters=10)
    t0 = time()
    clustering.fit(X_red)
    print("%s :\t%.2fs" % (linkage, time() - t0))

    plot_clustering(X_red, clustering.labels_, "%s linkage" % linkage)


plt.show()
```

### Пример 3

```python
print(__doc__)

import time as time
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_swiss_roll

# #############################################################################
# Generate data (swiss roll dataset)
n_samples = 1500
noise = 0.05
X, _ = make_swiss_roll(n_samples, noise=noise)
# Make it thinner
X[:, 1] *= .5

# #############################################################################
# Compute clustering
print("Compute unstructured hierarchical clustering...")
st = time.time()
ward = AgglomerativeClustering(n_clusters=6, linkage='ward').fit(X)
elapsed_time = time.time() - st
label = ward.labels_
print("Elapsed time: %.2fs" % elapsed_time)
print("Number of points: %i" % label.size)

# #############################################################################
# Plot result
fig = plt.figure()
ax = p3.Axes3D(fig)
ax.view_init(7, -80)
for l in np.unique(label):
    ax.scatter(X[label == l, 0], X[label == l, 1], X[label == l, 2],
               color=plt.cm.jet(float(l) / np.max(label + 1)),
               s=20, edgecolor='k')
plt.title('Without connectivity constraints (time %.2fs)' % elapsed_time)


# #############################################################################
# Define the structure A of the data. Here a 10 nearest neighbors
from sklearn.neighbors import kneighbors_graph
connectivity = kneighbors_graph(X, n_neighbors=10, include_self=False)

# #############################################################################
# Compute clustering
print("Compute structured hierarchical clustering...")
st = time.time()
ward = AgglomerativeClustering(n_clusters=6, connectivity=connectivity,
                               linkage='ward').fit(X)
elapsed_time = time.time() - st
label = ward.labels_
print("Elapsed time: %.2fs" % elapsed_time)
print("Number of points: %i" % label.size)

# #############################################################################
# Plot result
fig = plt.figure()
ax = p3.Axes3D(fig)
ax.view_init(7, -80)
for l in np.unique(label):
    ax.scatter(X[label == l, 0], X[label == l, 1], X[label == l, 2],
               color=plt.cm.jet(float(l) / np.max(label + 1)),
               s=20, edgecolor='k')
plt.title('With connectivity constraints (time %.2fs)' % elapsed_time)

plt.show()
```

### Пример 4

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances

np.random.seed(0)

# Generate waveform data
n_features = 2000
t = np.pi * np.linspace(0, 1, n_features)


def sqr(x):
    return np.sign(np.cos(x))

X = list()
y = list()
for i, (phi, a) in enumerate([(.5, .15), (.5, .6), (.3, .2)]):
    for _ in range(30):
        phase_noise = .01 * np.random.normal()
        amplitude_noise = .04 * np.random.normal()
        additional_noise = 1 - 2 * np.random.rand(n_features)
        # Make the noise sparse
        additional_noise[np.abs(additional_noise) < .997] = 0

        X.append(12 * ((a + amplitude_noise)
                 * (sqr(6 * (t + phi + phase_noise)))
                 + additional_noise))
        y.append(i)

X = np.array(X)
y = np.array(y)

n_clusters = 3

labels = ('Waveform 1', 'Waveform 2', 'Waveform 3')

# Plot the ground-truth labelling
plt.figure()
plt.axes([0, 0, 1, 1])
for l, c, n in zip(range(n_clusters), 'rgb',
                   labels):
    lines = plt.plot(X[y == l].T, c=c, alpha=.5)
    lines[0].set_label(n)

plt.legend(loc='best')

plt.axis('tight')
plt.axis('off')
plt.suptitle("Ground truth", size=20)


# Plot the distances
for index, metric in enumerate(["cosine", "euclidean", "cityblock"]):
    avg_dist = np.zeros((n_clusters, n_clusters))
    plt.figure(figsize=(5, 4.5))
    for i in range(n_clusters):
        for j in range(n_clusters):
            avg_dist[i, j] = pairwise_distances(X[y == i], X[y == j],
                                                metric=metric).mean()
    avg_dist /= avg_dist.max()
    for i in range(n_clusters):
        for j in range(n_clusters):
            plt.text(i, j, '%5.3f' % avg_dist[i, j],
                     verticalalignment='center',
                     horizontalalignment='center')

    plt.imshow(avg_dist, interpolation='nearest', cmap=plt.cm.gnuplot2,
               vmin=0)
    plt.xticks(range(n_clusters), labels, rotation=45)
    plt.yticks(range(n_clusters), labels)
    plt.colorbar()
    plt.suptitle("Interclass %s distances" % metric, size=18)
    plt.tight_layout()


# Plot clustering results
for index, metric in enumerate(["cosine", "euclidean", "cityblock"]):
    model = AgglomerativeClustering(n_clusters=n_clusters,
                                    linkage="average", affinity=metric)
    model.fit(X)
    plt.figure()
    plt.axes([0, 0, 1, 1])
    for l, c in zip(np.arange(model.n_clusters), 'rgbk'):
        plt.plot(X[model.labels_ == l].T, c=c, alpha=.5)
    plt.axis('tight')
    plt.axis('off')
    plt.suptitle("AgglomerativeClustering(affinity=%s)" % metric, size=20)


plt.show()
```
